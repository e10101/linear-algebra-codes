## Excercise 4-1

```python
def cosine_similarity(x, y):
    numerator = np.dot(x, y)
    # print('numerator', numerator)
    denominator = np.linalg.norm(x) * np.linalg.norm(y)
    # print('denominator', denominator)

    return numerator / denominator

def correlation_coefficient(x, y):
    xm = x - np.mean(x)
    ym = y - np.mean(y)

    numerator = np.dot(xm, ym)
    denominator = np.linalg.norm(xm) * np.linalg.norm(ym)

    return numerator / denominator
```

```python
a = np.random.randn(15)
b = np.random.randn(15)
# print(a, b)
print(cosine_similarity(a, b), 1 - distance.cosine(a, b))
print(correlation_coefficient(a, b), np.corrcoef(a, b)[0, 1])
```

```python
a, b = np.arange(0, 4), np.arange(-50, 51)
a, b
```

```python
a = np.expand_dims(a, axis=0)
b = np.expand_dims(b, axis=1)
```

```python
c = a + b
```

```python
c[:4, :]
```

```python
res_list = np.array([])
for idx in range(len(c)):
    row = c[idx, :]

    cs = cosine_similarity(a, row)
    cc = correlation_coefficient(a, row)
    res_list = np.append(res_list, [np.array([row[0]]), cs, cc])
```

```python
res_list = res_list.reshape((-1, 3))
```

```python
plt.plot(res_list[:, 0], res_list[:, 2], 'bo-')
plt.plot(res_list[:, 0], res_list[:, 1], 'gv-')

plt.xlabel('Mean offset')
plt.ylabel(' r or c')
plt.legend(['Person', 'Cosine sim'])
```
